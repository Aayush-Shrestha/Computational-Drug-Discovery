```python
selected_target = 'CHEMBL2051'
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
     
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(3)
```

```python
df.standard_type.unique()
```

```python

df2 = df[df.standard_value.notna()]
df2
```

```python

bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python

mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

```python

canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

```python

standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

```python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
     

df3
```

```python
pd.concat([df3,pd.Series(bioactivity_class)], axis=1)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem

def smiles_to_pdb(smiles, output_pdb):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        mol = Chem.AddHs(mol)
        AllChem.EmbedMolecule(mol, randomSeed=42)
        Chem.MolToPDBFile(mol, output_pdb)
    else:
        print("Invalid SMILES.")

if __name__ == "__main__":
    input_smiles = "CCC(CC)O[C@@H]1C=C(C(=O)O)C[C@H](N)[C@H]1NC(C)=O"  # Replace with your SMILES string
    output_pdb = "output1.pdb"  # Specify the output PDB file name
    
    smiles_to_pdb(input_smiles, output_pdb)
```
