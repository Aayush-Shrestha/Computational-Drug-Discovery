```python
import os
import shutil
import random

# Source directory and two destination directories
source_dir = '2'
destination_dir_1 = 'train/2'  # 75% of photos
destination_dir_2 = 'test/2'  # 25% of photos

# Get a list of all files in the source directory
all_photos = [f for f in os.listdir(source_dir) if os.path.isfile(os.path.join(source_dir, f))]

# Calculate the number of photos for each destination
num_photos_total = len(all_photos)
num_photos_dest_1 = int(num_photos_total * 0.75)
num_photos_dest_2 = num_photos_total - num_photos_dest_1

# Randomly select photos for each destination
photos_dest_1 = random.sample(all_photos, num_photos_dest_1)
photos_dest_2 = [photo for photo in all_photos if photo not in photos_dest_1]

# Function to copy photos to the respective destination directory
def copy_photos(photos, destination):
    for photo in photos:
        source_path = os.path.join(source_dir, photo)
        destination_path = os.path.join(destination, photo)
        shutil.copyfile(source_path, destination_path)
        print(f"Copied: {photo} to {destination_path}")

# Copy photos to destination directory 1 (75%)
copy_photos(photos_dest_1, destination_dir_1)

# Copy photos to destination directory 2 (25%)
copy_photos(photos_dest_2, destination_dir_2)
```
